```python
from transformers import BertTokenizer, BertModel
```

```python
# 加载BERT模型和分词器
tokenizer = BertTokenizer.from_pretrained('bert-base-chinese')
model = BertModel.from_pretrained('bert-base-chinese')


def sentence_similarity(sentence1, sentence2):
    # 对两个句子进行分词和编码
    encoded_inputs1 = tokenizer.encode_plus(sentence1,add_special_tokens=True, max_length=512, padding=True, truncation=True, return_tensors='pt')


    encoded_inputs2 = tokenizer.encode_plus(sentence2,add_special_tokens=True, max_length=512, padding=True, truncation=True, return_tensors='pt')
    
    # 计算BERT模型的输出
    with torch.no_grad():
        outputs1 = model(**encoded_inputs1)
        embeddings1 = outputs1.last_hidden_state.mean(dim=1)


    with torch.no_grad():
        outputs2 = model(**encoded_inputs2)
        embeddings2 = outputs2.last_hidden_state.mean(dim=1)
    
    
    # 计算余弦相似度得分
    score = torch.cosine_similarity(embeddings1[0], embeddings2[0], dim=0).item()
    
    return score
```

```python
test_sentence1 = "明天成都的天气怎么样"
test_sentence2 = "能告诉我明天在成都的天气吗"
test_sentence3 = "播放音乐"

print(sentence_similarity(test_sentence1,test_sentence2))
print(sentence_similarity(test_sentence1,test_sentence3))
print(sentence_similarity(test_sentence2,test_sentence3))
```
